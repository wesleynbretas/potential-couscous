```python
from prefect import flow, get_run_logger
```

```python
from tasks import (
    download_data,
    parse_data,
    save_report,
)
```

```python
@flow(name="process_data")
def process_data(n_users):
    logger = get_run_logger()
    
    # Tasks
    data = download_data(n_users)
    logger.info("Dados baixados com sucesso!")
    dataframe = parse_data(data)
    logger.info("Dados convertidos em Dataframe com sucesso!")
    save_report(dataframe)
    logger.info("Dados salvos em report.csv com sucesso!")
```

```python
if __name__ == "__main__":
    process_data(10)
```
